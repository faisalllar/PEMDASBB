```python
import pandas as pd  

# Mengatur tampilan agar semua baris muncul (opsional)
pd.set_option('display.max_rows', None)

# Membaca data dari file CSV
data = pd.read_csv('ds.csv')

# Menentukan tahun tertentu (misal: 2023)
tahun_tertentu = 2020

# Filter data berdasarkan tahun tertentu
data_tahun_tertentu = data[data['tahun'] == tahun_tertentu]

# Menghitung total produksi sampah untuk tahun tersebut
total_sampah = data_tahun_tertentu['jumlah_produksi_sampah'].sum()

# Menampilkan hasil
print(f"Total produksi sampah di Jawa Barat pada tahun {tahun_tertentu} adalah {total_sampah} ton.")
```

```python
print(data.columns)
```

```python
# Menampilkan semua kolom dalam loop
for column in data.columns:
    print(f"Kolom: {column}")
    print(data[column])
    print("-" * 30)
```
